```python
# Import dependencies
import os, requests, json
from config import api_key
```

```python
# Set up URL for test
city = 'New York City'
header = {'Authorization':f'Bearer {api_key}'}

url = f'https://api.yelp.com/v3/businesses/search?limit=50&location={city}'

print(url)
```

```python
# Make test request and show results
response = requests.get(url, headers = header).json()

print(json.dumps(response, indent=2, sort_keys=True))
```

```python
print(len(response['businesses']))
```

```python
for i in range(len(response['businesses'])):
    print(f"Business #{i}")
    if response['businesses'][i]['is_closed'] == 'true':
        print(f"{response['businesses'][i]['name']} is closed.")
    else:
        print(f"{response['businesses'][i]['name']} is open.")
```
