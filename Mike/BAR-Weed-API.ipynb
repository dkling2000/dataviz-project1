```python
#Sample usage of the program:
#`IMPORT DATA
from config import api_key
import json
import pprint
import requests
import sys
import pandas as pd

#load CSV File
city_file = "2018happycityrating.csv"


# Read file
happy_cities = pd.read_csv(city_file)
```

```python
#view import data
happy_cities.head()
```

# ## IMPORTING BAR DATA FROM YELP

```python
#IMPORT BAR DATA FROM YELP API
bar_data = []

for index, row in happy_cities.iterrows():
    city = row["City/State"]

    categories = "bars"
    header = {'Authorization':f'Bearer {api_key}'}
    radius = '16093'
    
    query_url = f'https://api.yelp.com/v3/businesses/search?limit=50&location={city}&categories={categories}&radius={radius}'
    bar_response = requests.get(query_url, headers = header)
    bar_json = bar_response.json()
    
    for i in range(len(bar_json['businesses'])):
        try:
            
            data = {
               
               'Name':bar_json['businesses'][i]['name'],
               'Price':bar_json['businesses'][i]['price'],
               'Rating':bar_json['businesses'][i]['rating'],
               'Review Count':bar_json['businesses'][i]['review_count'],
               'Out of Business':bar_json['businesses'][i]['is_closed'],
               'City':bar_json['businesses'][i]['location']['city'],
               'State':bar_json['businesses'][i]['location']['state'],
               'Distance':bar_json['businesses'][i]['distance'],
               'Address':bar_json['businesses'][i]['location']['address1']
                             
                        }
            
            # Add category data to dict
            for j in range(len(bar_json['businesses'][i]['categories'])):
                data[f'Category {j}'] = bar_json['businesses'][i]['categories'][j]['title']
                    
              # Add data to list
            bar_data.append(data)
        except KeyError:
            print('KeyError: skipping')
bars_df = pd.DataFrame(bar_data)
bars_df.to_csv("bars_data.csv", encoding="utf-8", index=False)
bars_df.head()
```
