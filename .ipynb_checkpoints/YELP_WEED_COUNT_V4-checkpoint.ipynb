```python
#Sample usage of the program:
#`IMPORT DATA
from config import api_key
import json
import pprint
import requests
import sys
import pandas as pd

#load CSV File
city_file = "2018happycityrating.csv"
bar_file = "bars_data.csv"
weed_file = "weed_data.csv"

# Read file
happy_cities = pd.read_csv(city_file)
bar_cities = pd.read_csv(bar_file)
weed_cities = pd.read_csv(weed_file)
happy_cities = happy_cities.rename(columns={"City/State": "Location"})
```

```python
#view import data
happy_cities.head()
```

```python
weed_cities.head()
```

```python
#IMPORT BAR DATA FROM YELP API
bar_data = []

for index, row in happy_cities.iterrows():
    city = row["City"]

    categories = "bars"
    header = {'Authorization':f'Bearer {api_key}'}
    radius = '8046'
    
    query_url = f'https://api.yelp.com/v3/businesses/search?limit=50&location={city}&categories={categories}&radius={radius}'
    bar_response = requests.get(query_url, headers = header)
    bar_json = bar_response.json()
    
    for i in range(len(bar_json['businesses'])):
        try:
            
            data = {
               
               'Name':bar_json['businesses'][i]['name'],
               'Price':bar_json['businesses'][i]['price'],
               'Rating':bar_json['businesses'][i]['rating'],
               'Review Count':bar_json['businesses'][i]['review_count'],
               'Out of Business':bar_json['businesses'][i]['is_closed'],
               'City':bar_json['businesses'][i]['location']['city'],
               'State':bar_json['businesses'][i]['location']['state'],
               'Distance':bar_json['businesses'][i]['distance'],
               'Address':bar_json['businesses'][i]['location']['address1']
                             
                        }
            
            # Add category data to dict
            for j in range(len(bar_json['businesses'][i]['categories'])):
                data[f'Category {j}'] = bar_json['businesses'][i]['categories'][j]['title']
                    
              # Add data to list
            bar_data.append(data)
        except KeyError:
            print('KeyError: skipping')
bars_df = pd.DataFrame(bar_data)
bars_df.to_csv("bars_data.csv", encoding="utf-8", index=False)
bars_df.head()
```

```python
#Creating Summary DF for Bar Data by City
bar_cities_group = bar_cities.groupby('City')
cities_rating_avg = bar_cities_group['Rating'].mean()
city_review_count = bar_cities_group['Review Count'].sum()
city_bar_count = bar_cities_group['Name'].count()

bar_summary = pd.DataFrame({"Bars Average Yelp Rating":cities_rating_avg, "Bars Count of Reviews":city_review_count, 
                            "Bar Count":city_bar_count})
bar_summary.head()
```

```python
#YELP API CALL FOR WEED DATA
weed_data = []

for index, row in happy_cities.iterrows():
    city = row["City"]

    categories = "cannabiscollective,cannabisdispensaries,cannabis_clinics,cannabistours,cannabisreferrals"
    header = {'Authorization':f'Bearer {api_key}'}
    radius = '16093'
    
    query_url = f'https://api.yelp.com/v3/businesses/search?limit=50&location={city}&categories={categories}&radius={radius}'
    weed_response = requests.get(query_url, headers = header)
    weed_json = weed_response.json()
    
    for i in range(len(weed_json['businesses'])):
        try:
            
            data = {
               
               'Name':weed_json['businesses'][i]['name'],
               'Price':weed_json['businesses'][i]['price'],
               'Rating':weed_json['businesses'][i]['rating'],
               'Review Count':weed_json['businesses'][i]['review_count'],
               'Out of Business':weed_json['businesses'][i]['is_closed'],
               'City':weed_json['businesses'][i]['location']['city'],
               'State':weed_json['businesses'][i]['location']['state'],
               'Distance':bar_json['businesses'][i]['distance'],
               'Address':weed_json['businesses'][i]['location']['address1']
                             
                        }
            
            # Add category data to dict
            for j in range(len(weed_json['businesses'][i]['categories'])):
                data[f'Category {j}'] = weed_json['businesses'][i]['categories'][j]['title']
                    
              # Add data to list
            weed_data.append(data)
        except KeyError:
            print('KeyError: skipping')
weed_df = pd.DataFrame(weed_data)
weed_df.to_csv("weed_data.csv", encoding="utf-8", index=False)
weed_df.head()
```

```python
weed_cities_group = weed_cities.groupby('City')
wcities_rating_avg = weed_cities_group['Rating'].mean()
wcity_review_count = weed_cities_group['Review Count'].sum()
wcity_bar_count = weed_cities_group['Name'].count()
weed_summary = pd.DataFrame({"Cannabis Shop Average Yelp Rating":wcities_rating_avg, "Cannabis Shop Count of Reviews":wcity_review_count, 
                            "Cannabis Shop Count":wcity_bar_count})
weed_summary.head()
```

```python
#Merging Happy Cities and Bar Summary
# Merge two dataframes using an outer join
merge_city_bar = pd.merge(happy_cities, bar_summary, on="City", how="left") 
merge_city_bar
```

```python
#Merging Happy Cities and Weed Summary
# Merge two dataframes using an outer join
merge_table_final = pd.merge(merge_city_bar, weed_summary, on="City", how="left") 
merge_table_final
```

```python
merge_table_final.to_csv("merge_table_final.csv", encoding="utf-8", index=False)
```
